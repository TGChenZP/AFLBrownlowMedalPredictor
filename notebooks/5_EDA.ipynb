```python
columns = data.columns

count_null = list()
for col in data.columns:
    count_null.append(data[col].isna().sum())

IDE = pd.DataFrame({'Number of Missing': count_null}, index = columns)
IDE
```

```python
mean = list()
std = list()
min = list()
q1 = list()
q2 = list()
q3 = list()
max = list()

for col in data.columns:
    if data[col].dtype in [np.float64, np.int64]:
        describe = data[col].describe()
        
        mean.append(describe['mean'])
        std.append(describe['std'])
        min.append(describe['min'])
        q1.append(describe['25%'])
        q2.append(describe['50%'])
        q3.append(describe['75%'])
        max.append(describe['max'])
    
    else:
        mean.append(np.nan)
        std.append(np.nan)
        min.append(np.nan)
        q1.append(np.nan)
        q2.append(np.nan)
        q3.append(np.nan)
        max.append(np.nan)

IDE['mean'] = mean
IDE['std'] = std
IDE['min'] = min
IDE['q1'] = q1
IDE['q2'] = q2
IDE['q3'] = q3
IDE['max'] = max

IDE['IQR'] = IDE['q3'] - IDE['q1']
IDE['1.5 upper bound'] = IDE['q3'] + 1.5*IDE['IQR']
IDE['1.5 lower bound'] = IDE['q1'] - 1.5*IDE['IQR']
IDE['3 upper bound'] = IDE['q3'] + 3*IDE['IQR']
IDE['3 lower bound'] = IDE['q1'] - 3*IDE['IQR']

outlier1 = list()
outlier3 = list()

for col in data.columns:
    if data[col].dtype in [np.float64, np.int64]:
        outlier1.append(len(data[(data[col] > IDE.loc[col]['1.5 upper bound']) | (data[col] < IDE.loc[col]['1.5 lower bound'])]))
        outlier3.append(len(data[(data[col] > IDE.loc[col]['3 upper bound']) | (data[col] < IDE.loc[col]['3 lower bound'])]))

    else:
        outlier1.append(np.nan)
        outlier3.append(np.nan)

IDE['number of 1.5 outliers'] = outlier1
IDE['number of 3 outliers']  = outlier3

IDE
```
